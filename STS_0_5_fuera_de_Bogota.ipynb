```python
def mapa_de_todos(df):
  temp_locations = df[['latitudes','longitudes']]
  temp_locationlist = temp_locations.values.tolist()
  avg_latitud = df['latitudes'].mean()
  avg_longitud = df['longitudes'].mean()
  avg_location = [avg_latitud, avg_longitud]
  map_todos = folium.Map(location=avg_location, zoom_start=12)
  folium.plugins.MarkerCluster(locations=temp_locationlist, overlay=True, popups=df['Marker']).add_to(map_todos)
  return map_todos
```

```python
mapa_de_todos(df)
```

```python
def mapa_de_ruta_actual(ruta):
  temp_df = df[df['RUTA']== ruta].reset_index()
  temp_locations =  temp_df[['latitudes', 'longitudes']]
  temp_locationlist = temp_locations.values.tolist()
  avg_latitud = temp_df['latitudes'].mean()
  avg_longitud = temp_df['longitudes'].mean()
  avg_location = [avg_latitud, avg_longitud]
  map = folium.Map(location=avg_location, zoom_start=14)
  folium.plugins.MarkerCluster(locations=temp_locationlist, overlay=True, popups=temp_df['Marker']).add_to(map)
  #for point in range(0, len(temp_locationlist)):
  #  folium.Marker(temp_locationlist[point], popup=temp_df['Direccion_final'][point]).add_to(map)
  return map
```

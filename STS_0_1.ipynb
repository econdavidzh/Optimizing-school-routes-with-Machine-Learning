```python
import pandas as pd
import numpy as np
import folium
import requests
```

```python
API_KEY = ''
direccion = 'Palo Alto, CA'
params = {'key':API_KEY, 
          'address': direccion
          }
base_url = 'https://maps.googleapis.com/maps/api/geocode/json?'
response = requests.get(base_url, params = params)
```

```python
df = pd.read_excel('/content/sts AI project.xlsx')
df['residence'] = df['Residencia'].astype(str) + ' Bogotá, Colombia'
residence = df["residence"].to_numpy()
```

```python
latitudes = []
longitudes = []
```

```python
for i in residence:
  direccion = i
  params = {'key':API_KEY, 
          'address': direccion}
  response = requests.get(base_url, params = params)
  #response.json()
  try:
    latitud = response.json()['results'][0]['geometry']['location']['lat']
    longitud = response.json()['results'][0]['geometry']['location']['lng']
    print(i, latitud, longitud)
    latitudes.append(latitud)
    longitudes.append(longitud)
  except:
    print(i + " not found")
    latitudes.append('not found')
    longitudes.append('not found')
```

```python
df['latitudes'] = latitudes
df['longitudes'] = longitudes
df['coordenadas']= list(zip(df.latitudes, df.longitudes))
```

```python
df_1 = df[df['RUTA ']==1]
df_2 = df[df['RUTA ']==2]
df_5 = df[df['RUTA ']==5]
df_6 = df[df['RUTA ']==6]
df_25 = df[df['RUTA ']==25]
```

```python
dict_of_locations_1 = pd.Series(df_1.coordenadas.values,index=df_1.Alumno).to_dict()
dict_of_locations_2 = pd.Series(df_2.coordenadas.values,index=df_2.Alumno).to_dict()
dict_of_locations_5 = pd.Series(df_5.coordenadas.values,index=df_5.Alumno).to_dict()
dict_of_locations_6 = pd.Series(df_6.coordenadas.values,index=df_6.Alumno).to_dict()
dict_of_locations_25 = pd.Series(df_25.coordenadas.values,index=df_25.Alumno).to_dict()
```

```python
# Setting up Map:
STS = [4.710664, -74.124992]
mapa_de_alumnos = folium.Map(location=STS, zoom_start=12, tiles='OpenStreetMap', width='90%', height='90%')
```

```python
# Adding locations to Map:
for i in dict_of_locations_1:
  folium.Marker(location = dict_of_locations_1[i], tooltip=i).add_to(mapa_de_alumnos)
for i in dict_of_locations_2:
  folium.Marker(location = dict_of_locations_2[i], tooltip=i, icon=folium.Icon(color="red", icon="info-sign")).add_to(mapa_de_alumnos)
for i in dict_of_locations_5:
  folium.Marker(location = dict_of_locations_5[i], tooltip=i, icon=folium.Icon(color="green", icon="info-sign")).add_to(mapa_de_alumnos)
for i in dict_of_locations_6:
  folium.Marker(location = dict_of_locations_6[i], tooltip=i, icon=folium.Icon(color="black")).add_to(mapa_de_alumnos)
for i in dict_of_locations_25:
  folium.Marker(location = dict_of_locations_25[i], tooltip=i, icon=folium.Icon(color="pink", icon="info-sign")).add_to(mapa_de_alumnos)
```
